```python
#cleaning columns from spaces and unnecessary characters
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]+', '', regex=True)
```

```python
#renaming columns for consistency
df.rename({'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_children'}, axis=1, inplace=True)
```

```python
df.columns
```

```python
#filling missing value cells with appropriate values
df['agent'].fillna(-1, inplace=True)
df['country'].fillna('unknown', inplace=True)
df[['agent', 'country']].isna().sum()
```

```python
df.isna().sum()
```

```python
#removing columns with highest missing values and least effectiveness on the dataset
df.drop(columns=['company'], inplace=True)
```

```python
df.isna().sum()
```

```python
#understanding datatypes
df.dtypes.to_dict
```

```python
df['num_children'] = df['num_children'].fillna(0)
df['agent'] = df['agent'].fillna(0)
```

```python
df = df.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int64', 'agent': 'int64'})
df.dtypes
```

```python
df.isna().sum()
```

```python
#count duplicates and delete them
df.duplicated().sum()   

df = df.drop_duplicates()
```

```python
df.duplicated().sum() 
```

```python
#changing Average Day Rating type and detecting outliers
df['adr'] = df['adr'].astype('int64')
```

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_adr = df[(df['adr'] < lower_bound) | (df['adr'] > upper_bound)]
outliers_adr.shape
```

```python
#filtering out extreme outliers
df = df[(df['adr'] >= lower_bound) & (df['adr'] <= upper_bound)]
```

```python
#finding duplicate columns
df.loc[:, df.columns.duplicated()]
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
#final check out to make sure it is perfectly cleaned
df.info()
df.isna().sum()
df.describe()
df.head()
df.duplicated().sum()
```
